## Складання аналітичної моделі

Запишемо в загальному вигляді диференціальне рівняння перехідного процесу для посудини. Отже, за елементарний проміжок часу для посудини рівняння масового балансу має вигляд:

Запишемо в загальному вигляді диференціальне рівняння перехідного процесу для посудини. Отже, за елементарний проміжок часу   для посудини рівняння масового балансу має вигляд:

$$\rho · F · dz(t) = G_{in1} (t)· dt + G_{in2} (t)· dt - G_{out1}(t) · dt - G_{out2} · dt - G_{out3} · dt$$

або

$$\rho · F · \frac {dz(t)}{dt} = G_{in1} (t) + G_{in2} (t) - G_{out1}(t) - G_{out2} - G_{out3}$$

де  $\rho$ - густина рідини, $кг/м^3$ ;

$F$ - площа поперечного перерізу посудини, $м^2$
 
$$F = \frac{\pi · D^2}{4}$$


$D$ - діаметр поперечного перерізу посудини, $м$;

$z(t)$ - рівень води в посудині, м;

$$z_{1}(t) = z(t) - (z_{0}-z_{1,0})$$
$$z_{2}(t) = z(t) - (z_{0}-z_{2,0})$$


$G_{in1}(t)$, $G_{in2}(t)$, $G_{out1}(t)$, $G_{out2}(t)$, $G_{out3}(t)$  - масові витрати, $кг/сек$;

$$G_{in1}(t) = \rho · a_{in1}(t) · x_{in1}(t) · \sqrt{p_{in1}(t)}$$
$$G_{in2}(t) = \rho · a_{in2}(t) · x_{in2}(t) · \sqrt{p_{in2}(t) - z_{1}(t)}$$
$$G_{out1}(t) = \rho · a_{out1}(t) · x_{out1}(t) · \sqrt{z(t) - p_{out1}(t)}$$
$$G_{out2}(t) = \rho · a_{out2}(t) · x_{out2}(t) · \sqrt{z(t) - p_{out2}(t)}$$
$$G_{out3}(t) = \rho · a_{out3}(t) · x_{out3}(t) · \sqrt{z_{2}(t) - p_{out3}(t)}$$

$a_{in1}(t)$, $a_{in2}(t)$, $a_{out1}(t)$, $a_{out2}(t)$ $a_{out3}(t)$ - коефіцієнти витрати, $м^{0,5}/сек$;
$p_{in}$, $p_{out1}$, $p_{out2}$ - надлишкові тиски, $м$.

$x_{in1}(t)$, $x_{in2}(t)$, $x_{out1}(t)$, $x_{out2}(t)$, $x_{out3}(t)$ - площі перерізів для проходу рідини через вентилі, $м^2$, які за умови круглої форми визначаються як:

$$x_{in1}(t) = \frac{\pi · d^2_{in1}}{4}$$
$$x_{in2}(t) = \frac{\pi · d^2_{in2}}{4}$$
$$x_{out1}(t) = \frac{\pi · d^2_{out1}}{4}$$
$$x_{out2} = \frac{\pi · d^2_{out2,0}}{4}$$
$$x_{out3} = \frac{\pi · d^2_{out3,0}}{4}$$

$d_{in1}(t)$, $d_{in2}(t)$, $d_{out1}(t)$, $d_{out2}$,  $d_{out3}$ - діаметри перерізів для проходу рідини через вентилі, $м$;
 

Врахувавши, що $p$, $a$, $d_{out2}$ , $d_{out3}$ – не змінюються в часі, рівняння масового балансу можна записати у вигляді:

$$ F · \frac {dz(t)}{dt} =  (\rho · a_{in1,0} · x_{in1}(t) · \sqrt{p_{in1,0}}) +  (\rho · a_{in2,0} · x_{in2}(t) · \sqrt{p_{in2,0} - z_{1}(t)}) - (\rho · a_{out1,0} · x_{out1}(t) · \sqrt{z(t) - p_{out1,0}}) - (\rho · a_{out2,0} · x_{out2,0} · \sqrt{z(t) - p_{out2,0}}) - (\rho · a_{out3,0} · x_{out3,0} · \sqrt{z_{2}(t) - p_{out3,0} })$$

де індекс $0$ добавлено для позначення параметрів, що не змінюються у перехідному процесі.

В даному прикладі не задано коефіцієнт витрати вихідного вентиля 3. Використовуючи залежність:

$$G_{in1,0} + G_{in2,0} = G_{out1,0} + G_{out2,0} + G_{out3,0}$$

де  $G_{in1,0}, G_{in2,0}, G_{out1,0}, G_{out2,0}, G_{out3,0}$ - масові витрати ($0$ добавлено для позначення параметрів у стаціонарному режимі або незмінності параметрів під час перехідного процесу), кг/сек


$$ G_{in1}(t) =  \rho · a_{in1,0} · x_{in1}(t) · \sqrt{p_{in1,0}}$$
$$ G_{in2}(t) =  \rho · a_{in2,0} · x_{in2}(t) · \sqrt{p_{in2,0} - z_{1}(t)}$$
$$ G_{out1}(t) = \rho · a_{out1,0} · x_{out1}(t) · \sqrt{z(t) - p_{out1,0}}$$
$$ G_{out2}(t) = \rho · a_{out2,0} · x_{out2,0} · \sqrt{z(t) - p_{out2,0}}$$
$$ G_{out3}(t) = \rho · a_{out3,0} · x_{out3,0} · \sqrt{z_{2}(t) - p_{out3,0}}$$

Таким чином рівняння масового балансу можгна записати у вигляді:

$$ (a_{in1,0} · x_{in1}(t) · \sqrt{p_{in1,0}}) + (a_{in2,0} · x_{in2}(t) · \sqrt{p_{in2,0} - z_{1}(t)}) = (a_{out1,0} · x_{out1}(t) · \sqrt{z(t) - p_{out1,0}}) + (a_{out2,0} · x_{out2,0} · \sqrt{z(t) - p_{out2,0}}) + (a_{out3,0} · x_{out3,0} · \sqrt{z_{2}(t) - p_{out3,0} })$$

З даного рівняння можна вивести значення коефіцієнта витрати вихідного вентиля:

$$ a_{out3,0} = \frac{(a_{in1,0} · \frac{\pi · d^2_{in1}}{4} · \sqrt{p_{in1,0}}) + (a_{in2,0} · \frac{\pi · d^2_{in2}}{4} · \sqrt{p_{in2,0} - (z(t) - (z_{0}-z_{1,0}))} - (a_{out1,0} · \frac{\pi · d^2_{out1}}{4} · \sqrt{z(t) - p_{out1,0}}) - (a_{out2,0} · \frac{\pi · d^2_{out2,0}}{4} · \sqrt{z(t) - p_{out2,0}})}
{\frac{\pi · d^2_{out3,0}}{4} · \sqrt{(z(t) - (z_{0}-z_{2,0})) - p_{out3,0}}}
$$

Отримані диференціальні рівняння реалізуємо в вигляді комп'ютерної моделі на мові програмування Python.

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import math

D = 0.2

pin1_0 = 1
pin2_0 = 0.9
pout1_0 = 0.02
pout2_0 = 0.03
pout3_0 = 0.05

z0_0 = 0.6
z1_0 = 0.1
z2_0 = 0.15

din1_0 = 0.01
din2_0 = 0.04
dout1_0 = 0.025
dout2_0 = 0.03
dout3_0 = 0.03

ain1_0 = 0.01
ain2_0 = 0.02
aout1_0 = 0.01
aout2_0 = 0.02


F = (math.pi*D**2)/4

xin1_0 = (math.pi*din1_0**2)/4
xin2_0 = (math.pi*din2_0**2)/4
xout1_0 = (math.pi*dout1_0**2)/4
xout2_0 = (math.pi*dout2_0**2)/4
xout3_0 = (math.pi*dout3_0**2)/4


aout3_0 = (ain1_0 * xin1_0 * math.sqrt(pin1_0) +
ain2_0 * xin2_0 * math.sqrt(pin2_0 - z1_0) - 
aout1_0 * xout1_0 * math.sqrt(z0_0 - pout1_0) - 
aout2_0 * xout2_0 * math.sqrt(z0_0 - pout2_0) / 
xout3_0 * math.sqrt(z2_0 - pout3_0))

def Dynamic (z,t,u):

  xin1, xin2, xout1 = u
  z0 = z

  z1t = (z0 - (z0_0 - z1_0))
  z2t = (z0 - (z0_0 - z2_0))

  Gin1 = ain1_0 * xin1 * math.sqrt(pin1_0)
  Gin2 = ain2_0 * xin2 * math.sqrt(pin2_0 - z1t) if z0 <= (pin2_0 + z0_0 - z1_0) else -(ain2_0 * xin2 * math.sqrt(abs(pin2_0 - z1t)))
  Gout1 = aout1_0 * xout1 * math.sqrt(z0 - pout1_0) if z0 >= pout1_0 else 0
  Gout2 = aout2_0 * xout2_0 * math.sqrt(z0 - pout2_0) if z0 >= pout2_0 else 0
  Gout3 = aout3_0 * xout3_0 * math.sqrt(z2t - pout3_0) if z0 >= (z0_0 - z2_0 + pout3_0) else 0
  
  dz_dt = (Gin1 + Gin2 - Gout1 - Gout2 - Gout3) / F

  return dz_dt

# initial condition

# time points
t = np.linspace(0,9000, 10000)
u0 = [1.1 * xin1_0, 1.0 * xin2_0, 1.1 * xout1_0] #Задаємо збурення системі
z0 = 0
# solve ODE
y = odeint(Dynamic,z0,t,(u0,)).round(5)


# plot results
plt.plot(t,y)
plt.xlabel('time, sec')
plt.ylabel('z(t), m')
plt.show()
print(y)
```
